```python
from datasets import load_dataset
from huggingface_hub import HfApi
api = HfApi()

langs = ["az","sq","ta","sr","he","lv","sl","et","bn","id","ca","lt","th","sk","hi","no","bg","da","ko","fi","uk","sv","ro","el","hu","fa","vi","cs","tr","ar","nl","ja","pl","pt","it","zh","fr","de","es","ru","en"]
print(len(langs))

for lang in langs:

  filename = f"culturaX-{lang}.txt"
  dataset = load_dataset('uonlp/CulturaX',lang,split='train',streaming=True)
  print(dataset)

  c = 0
  with open(filename,'a') as f:
    for row in iter(dataset):
      if c % 100000 == 0 and c!=0:
        print(c)
      f.write(f'{row["text"]}<eot>')
      c += 1

  api.upload_file(
        path_or_fileobj="/content/"+filename,
        path_in_repo="pretrain/"+filename,
        repo_id="vatsadev/fusion",
        repo_type="dataset",
  )
```

```python
from datasets import load_dataset
from huggingface_hub import HfApi
api = HfApi()


filename = f"openMathnvidia.txt"
dataset = load_dataset('nvidia/OpenMathInstruct-1', split='train')
print(dataset)

c = 0
with open(filename,'a') as f:
  for row in iter(dataset):
      if c % 100000 == 0 and c!=0:
        print(c)
      f.write(f'Q {row["question"]} A{row["expected_answer"]}<eot>')
      c += 1

api.upload_file(
        path_or_fileobj="/content/"+filename,
        path_in_repo="/instruct/"+filename,
        repo_id="VatsaDev/fusion",
        repo_type="dataset",
  )
```
